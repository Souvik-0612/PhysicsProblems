```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import sin, cos, pi, sqrt, exp
plt.style.use(['science', 'notebook', 'grid'])
```

```python
def DtoR(x): return np.pi*x/180

def projectile(v, a, tf, h):
    ti = 0 #sec
    xi = 0; yi = 0; v_x = v*np.cos(DtoR(a)); v_y = v*np.sin(DtoR(a)); 	
    xx = [xi]; yy = [yi]
    def v_xDE(t, x, v_x): return 0
    def v_yDE(t, y, v_y): return -9.8 #m/s^2
    while (tf > ti):
        ti += h
        v_x += h*v_xDE(ti, xi, v_x)
        xi += h*v_x
        v_y += h*v_yDE(ti, yi, v_y)
        yi += h*v_y
        if yi<0:
            break
        xx.append(xi)
        yy.append(yi)
    
    return xx, yy

plt.figure(figsize = (16, 12))
plt.title("Projectile at diffarent angles", size = 25)
for i in range(5, 90, 5):
    x, y = projectile(30, i, 10, 0.001)
    xx = np.array(x); yy = np.array(y)
    h = xx[1: ] - xx[0: -1]
    n = len(xx)
    s = h*(yy[0] + 4*sum(yy[1: n-1: 2]) + 2*sum(yy[2: n-1: 2]) + yy[n-1])/3
    print(f"For angle {i}° Area is {sum(s)/n:.3f}")
    plt.plot(x, y, label = str(i)+"°")
plt.legend(loc = "best")
plt.show()
```
