# Lorentz force


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
plt.style.use(['science', 'notebook', 'grid'])
```

## Griffiths examples

### Example 5.1: Cyclotron motion

```python
#Parameter 
q = 1; m = 1
Vx, Vy, Vz = (0, 1, 0)
Bx, By, Bz = (0, 0, -1)
Ex, Ey, Ez = (0, 0, 0)

def dSdt(S, t):
    x, Vx, y, Vy, z, Vz = S
    return [Vx,
           (q/m)*(Ex + Vy*Bz - Vz*By),
           Vy,
           (q/m)*(Ey + Vz*Bx - Vx*Bz),
           Vz,
           (q/m)*(Ez + Vx*By - Vy*Bx)]

S0 = 1, 0, 0, 1, 0, 0   #Initial position is (1, 0, 0) & Initial Velocity (0, 1, 0)
t = np.linspace(0, 10, 100)

Sol = odeint(dSdt, S0, t)

x, y, z = Sol.T[0], Sol.T[2], Sol.T[4]

#Plotting
plt.figure(figsize = (18, 18))
plt.title("Charged particles in both magnectic and Electric field")
plt.subplot(221)
plt.plot(x, y)
plt.xlabel("<--X-Axis-->")
plt.ylabel("<--Y-Axis-->")
plt.subplot(222)
plt.plot(y, z)
plt.xlabel("<--Y-Axis-->")
plt.ylabel("<--Z-Axis-->")
plt.subplot(223)
plt.plot(z, x)
plt.xlabel("<--Z-Axis-->")
plt.ylabel("<--X-Axis-->")

plt.show()
```

### Example 5.2: Cycloid motion

```python
#Parameter 
q = 1; m = 1
Vx, Vy, Vz = (1, 0, 0)
Bx, By, Bz = (1, 0, 0)
Ex, Ey, Ez = (0, 0, 1)

def dSdt(S, t):
    x, Vx, y, Vy, z, Vz = S
    return [Vx,
           (q/m)*(Ex + Vy*Bz - Vz*By),
           Vy,
           (q/m)*(Ey + Vz*Bx - Vx*Bz),
           Vz,
           (q/m)*(Ez + Vx*By - Vy*Bx)]

S0 = 0, 0, 0, 0, 0, 0 #Initial position is (0, 0, 0) & Initial Velocity (0, 0, 0)
t = np.linspace(0, 10, 100)

Sol = odeint(dSdt, S0, t)

x, y, z = Sol.T[0], Sol.T[2], Sol.T[4]

#Plotting
plt.figure(figsize = (18, 14))
plt.title("Charged particles in both magnectic and Electric field")
plt.subplot(221)
plt.plot(x, y)
plt.xlabel("<--X-Axis-->")
plt.ylabel("<--Y-Axis-->")
plt.subplot(222)
plt.plot(y, z)
plt.xlabel("<--Y-Axis-->")
plt.ylabel("<--Z-Axis-->")
plt.subplot(223)
plt.plot(z, x)
plt.xlabel("<--Z-Axis-->")
plt.ylabel("<--X-Axis-->")

plt.show()
```
